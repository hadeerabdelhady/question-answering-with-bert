```python
from datasets import load_dataset# datasets try load our dataset in object suitable for transformers
from transformers import  TFBertModel,BertTokenizerFast
import tensorflow as tf
```

```python
train_df.to_csv("train.csv",index=False)
test_df.to_csv("test.csv",index=False)
valid_df.to_csv("valid.csv",index=False)
```

```python
dataset=load_dataset("csv",data_files={'train':"dataset/train.csv",'test':"dataset/test.csv"})#convert my dataset into object suitable for transformers
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
def preprocess(example):
    if example["answer"] is None or example["answer_start"] is None:
        return None
    inputs = tokenizer(#tokenize question and context only
        example["question"],
        example["context"],
        max_length=384,
        truncation="only_second",
        padding="max_length",
        return_offsets_mapping=True
    )

    # Calculate start/end token positions for the answer
    start_char = example["answer_start"]# The starting character index of the answer in the context
    end_char = start_char + len(example["answer"]) # The ending character index of the answer in the context
    offsets = inputs["offset_mapping"]#Get the list of (start, end) offsets for each token

    # Find the token positions of the answer
    start_pos = end_pos = 0 # Initialize positions (start and end) to 0
    for idx, (start, end) in enumerate(offsets):
        if start <= start_char < end: # If the token contains the starting character of the answer
            start_pos = idx# Set the start position to the current token's index
        if start < end_char <= end:# If the token contains the ending character of the answer
            end_pos = idx # Set the end position to the current token's index
            break
# Add the start and end token positions to the tokenized input 
    inputs["start_positions"] = start_pos # Position of the answer's start token
    inputs["end_positions"] = end_pos# Position of the answer's end token
    inputs.pop("offset_mapping")  # No need to keep this during training , Remove the 'offset_mapping' as it's not needed for training

    return inputs# Return the processed inputs with the added start and end positions

#“Here is a question + context pair. Turn them into tokens, and tell me where each token came from in the original text (using offset_mapping).”
```

```python
tokenized_train = dataset["train"].map(preprocess)
tokenized_test = dataset["test"].map(preprocess)
```

```python
from transformers import BertForQuestionAnswering

# Load pre-trained BERT model with a QA head
model = BertForQuestionAnswering.from_pretrained("bert-base-uncased")# call model special for question answering tasks from large language model bert-base-uncased
```

after building model we want to train it with some trainig arguments

```python
from transformers import Trainer,TrainingArguments
training_arg=TrainingArguments(
    output_dir="dataset/results",# store our output in this directory
    eval_strategy="epoch",# we will follow strategy of ebochs 
    learning_rate=2e-5,
    per_device_train_batch_size=4,# four samples each time
    per_device_eval_batch_size=4,
    num_train_epochs=1,# number of ebochs 
    weight_decay=0.01,# intialization of weights 


)
#similar to compile process
trainer=Trainer(
    model=model,#the model that you determine it that is BertForQuestionAnswering
    args=training_arg,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,


)
```

```python
trainer.train()
```

```python
evaluate_result=trainer.evaluate()
print(evaluate_result['eval_loss'])
```
